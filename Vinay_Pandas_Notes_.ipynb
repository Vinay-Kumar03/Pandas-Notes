# Time series


```python
#what is time series data?
"""Time series data is a sequence of data points in cronological
order that is used by business to analyze the past data and make the future predictions"""
#weather forecast
#no of taxi rides in a month
#predicting a company stock price fo the next day
```

```python
#components of time series
# it is consists of 4 components
# 1. Trend component :- This is a variation that moves up or down in a predictable pettern over long period
# 2. sesonality component :- The variation that is regular and periodic and repeats itself over a specific period is known as sesonality component
  #ex:- such as days,weeks, or the months,seasons
# 3. cyclic component :- is the variation that corresponds with business or economic, "boom-bust" cycles it follows their own pecular
# 4. random component :- irregular component
```

# Pivot table

```python
df=pd.DataFrame(
    {
        "A":['one','one','two','three']*3,
        "B":["A","B","C"]*4,
        "C":['Food','Food','Starter','Drink','Icecream','pan']*2,
        "D":np.random.randint(0,12,size=12),
        "E":np.random.randint(0,12,size=12)
    }
)
df
```

```python
#pd.pivot_table()
pd.pivot_table(df,values='D',index=['A','B'],columns=['C'])
```

```python
df=pd.DataFrame(
    {
        "A":['one','one','two','three']*3,
        "B":["A","B","C"]*4,
        "C":['Food','Food','Starter','Drink','Icecream','pan']*2,
        "D":np.random.randint(0,12,size=12),
        "E":np.random.randint(0,12,size=12)
    }
)
df
```

```python
#group by function is applicable with the aggregate function
#min,max.sum,avg,count,median,var,std
```

```python
#print the sum of food and drink for "c" and "e"
df.groupby("C")[["D","E"]].std()
```
